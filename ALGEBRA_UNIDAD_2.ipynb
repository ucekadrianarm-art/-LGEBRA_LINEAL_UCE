## ¿Qué es una matriz?

En matemáticas, una **matriz** es un arreglo rectangular de números (o elementos de otro tipo) organizados en filas y columnas. Se utilizan para representar sistemas de ecuaciones lineales, transformaciones lineales y para almacenar datos en diversas aplicaciones, como en la computación gráfica, la física y la economía.

Las matrices se denotan generalmente con letras mayúsculas y sus elementos con letras minúsculas con subíndices que indican su posición (fila y columna). Por ejemplo, una matriz $A$ de tamaño $m \times n$ tiene $m$ filas y $n$ columnas, y su elemento en la fila $i$ y columna $j$ se denota como $a_{ij}$.

# Operaciones con matrices

## Operaciones con matrices (Teoría)

### Suma de matrices

Para sumar dos matrices, es necesario que tengan las mismas dimensiones (el mismo número de filas y columnas). La suma se realiza elemento a elemento, es decir, el elemento en la posición $(i, j)$ de la matriz resultante es la suma de los elementos en la misma posición de las matrices originales.

Si $A = [a_{ij}]$ y $B = [b_{ij}]$ son dos matrices de tamaño $m \times n$, entonces su suma $C = A + B$ es una matriz de tamaño $m \times n$ donde $c_{ij} = a_{ij} + b_{ij}$.

### Multiplicación por un escalar

La multiplicación de una matriz por un escalar implica multiplicar cada elemento de la matriz por ese escalar. Si $A = [a_{ij}]$ es una matriz de tamaño $m \times n$ y $k$ es un escalar, entonces la matriz resultante $B = kA$ es una matriz de tamaño $m \times n$ donde $b_{ij} = k \cdot a_{ij}$.

### Producto de matrices (Producto matricial)

El producto de dos matrices $A$ y $B$, denotado como $AB$, solo es posible si el número de columnas de la primera matriz ($A$) es igual al número de filas de la segunda matriz ($B$). Si $A$ es una matriz de tamaño $m \times n$ y $B$ es una matriz de tamaño $n \times p$, entonces el producto $C = AB$ es una matriz de tamaño $m \times p$.

El elemento en la posición $(i, j)$ de la matriz resultante $C$ se calcula como la suma de los productos de los elementos de la fila $i$ de la matriz $A$ por los elementos correspondientes de la columna $j$ de la matriz $B$. Formalmente:

$c_{ij} = \sum_{k=1}^{n} a_{ik} b_{kj}$

### Producto escalar (Producto punto)

El término "producto escalar" o "producto punto" puede referirse a diferentes operaciones dependiendo del contexto.

*   **Para vectores:** Es una operación que toma dos vectores y devuelve un escalar. Se calcula multiplicando los elementos correspondientes de los vectores y sumando los resultados. Para dos vectores $\mathbf{u} = [u_1, u_2, \dots, u_n]$ y $\mathbf{v} = [v_1, v_2, \dots, v_n]$, su producto escalar es $\mathbf{u} \cdot \mathbf{v} = \sum_{i=1}^{n} u_i v_i$.

*   **Para matrices (a veces):** En algunos contextos, se puede referir a la multiplicación elemento a elemento de dos matrices de las mismas dimensiones, similar a la suma, pero con multiplicación en lugar de suma. Sin embargo, el término más común para esto es **producto de Hadamard** o **producto elemento a elemento**. La operación `*` en NumPy entre dos matrices de las mismas dimensiones realiza este producto elemento a elemento.

## Definir matrices

Define dos matrices usando Python


```python
import numpy as np

# Definir dos matrices de ejemplo
matrix_a = np.array([[1, 2], [3, 4]])
matrix_b = np.array([[5, 6], [7, 8]])
scalar = 2

print("Matriz A:")
display(matrix_a)
print("\nMatriz B:")
display(matrix_b)
print("\nEscalar:")
display(scalar)

# Suma de matrices
if matrix_a.shape == matrix_b.shape:
  suma_matrices = matrix_a + matrix_b
  print("\nSuma de matrices (A + B):")
  display(suma_matrices)
else:
  print("\nNo se puede realizar la suma de matrices: las dimensiones no coinciden.")

# Multiplicación por escalar
multiplicacion_escalar = matrix_a * scalar
print("\nMultiplicación de la Matriz A por el escalar:")
display(multiplicacion_escalar)

# Producto escalar (producto punto)
# Para el producto escalar (producto punto) entre dos vectores, las dimensiones deben ser las mismas.
# Si se aplica a matrices, realiza el producto elemento a elemento.
if matrix_a.shape == matrix_b.shape:
  producto_escalar = np.dot(matrix_a.flatten(), matrix_b.flatten()) # Convierte a vectores para producto punto
  print("\nProducto escalar (producto punto) entre A y B (tratadas como vectores):")
  display(producto_escalar)
else:
    print("\nNo se puede realizar el producto escalar: las dimensiones no coinciden.")


# Producto matricial
# Para el producto matricial, el número de columnas de la primera matriz debe ser igual al número de filas de la segunda.
if matrix_a.shape[1] == matrix_b.shape[0]:
  producto_matricial = np.dot(matrix_a, matrix_b)
  print("\nProducto matricial (A * B):")
  display(producto_matricial)
else:
  print("\nNo se puede realizar el producto matricial: las dimensiones no coinciden.")
```
