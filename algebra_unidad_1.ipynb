# Operaciones con vectores


## Definir vectores

Define tres vectores usando Python


**Razonamiento:** Defina tres vectores 2D utilizando matrices NumPy como se solicita en las instrucciones.



```python
import numpy as np

# Define the three 2D vectors
vector1 = np.array([2, 3])
vector2 = np.array([-1, 5])
vector3 = np.array([4, -2])

# Display the vectors
print("Vector 1:", vector1)
print("Vector 2:", vector2)
print("Vector 3:", vector3)
```

## Operación con vectores

### Subtarea:

Realizar diversas operaciones vectoriales, como suma, resta, multiplicación escalar, producto escalar y magnitud.


**Razonamiento**:
Realice las operaciones vectoriales solicitadas: suma, diferencia, multiplicación escalar, producto escalar y magnitud.



```python
# 1. Calculate the sum of vector1 and vector2
vector_sum = vector1 + vector2
print("Suma de vector 1+vector 2:", vector_sum)

# 2. Calculate the difference between vector1 and vector3
vector_difference = vector1 - vector3
print("Diferencia entre el vector 1 y el vector 3:", vector_difference)

# 3. Calculate the scalar multiplication of vector2 by a scalar (e.g., 2)
scalar = 2
vector_scalar_multiplication = scalar * vector2
print("Multiplicación del escalar por el vector2 ", scalar, ":", vector_scalar_multiplication)

# 4. Calculate the dot product of vector1 and vector3
dot_product = np.dot(vector1, vector3)
print("Producto punto del vector 1 y el vector 3:", dot_product)

# 5. Calculate the magnitude of vector2
magnitude_vector2 = np.linalg.norm(vector2)
print("Magnitud del vector 2:", magnitude_vector2)
```


## Resultados

### Subtarea:
Muestra los resultados de las operaciones vectoriales.


**Razonamiento**:
Imprima los resultados de las operaciones vectoriales realizadas en el paso anterior con etiquetas descriptivas.


```python
print("Resultado de las operaciones vectoriales:")
print("Suma de vector1 y vector2:", vector_sum)
print("Diferencia entre vector1 y vector3:", vector_difference)
print("multiplicación del vector2 por un escalar", scalar, ":", vector_scalar_multiplication)
print("Producto punto vector1 y vector3:", dot_product)
print("Magnitud del vector2:", magnitude_vector2)
```

## Opcional: Visualizar vectores

### Subtarea:
Visualizar los vectores y sus operaciones.



**Razonamiento**:
Visualice los vectores y sus operaciones usando matplotlib.



```python
import matplotlib.pyplot as plt

# Create a figure and axes object
fig, ax = plt.subplots(figsize=(8, 8))

# Plot the initial vectors
ax.quiver(0, 0, vector1[0], vector1[1], angles='xy', scale_units='xy', scale=1, color='r', label='vector1')
ax.quiver(0, 0, vector2[0], vector2[1], angles='xy', scale_units='xy', scale=1, color='g', label='vector2')
ax.quiver(0, 0, vector3[0], vector3[1], angles='xy', scale_units='xy', scale=1, color='b', label='vector3')

# Plot the results of vector operations
# Vector sum (vector1 + vector2) - starting vector2 from the end of vector1
ax.quiver(0, 0, vector_sum[0], vector_sum[1], angles='xy', scale_units='xy', scale=1, color='purple', label='vector_sum')
ax.quiver(vector1[0], vector1[1], vector2[0], vector2[1], angles='xy', scale_units='xy', scale=1, color='orange', label='vector2 from end of vector1')


# Vector difference (vector1 - vector3) - starting the result from the origin
ax.quiver(0, 0, vector_difference[0], vector_difference[1], angles='xy', scale_units='xy', scale=1, color='brown', label='vector_difference')


# Scalar multiplication (scalar * vector2) - starting from the origin
ax.quiver(0, 0, vector_scalar_multiplication[0], vector_scalar_multiplication[1], angles='xy', scale_units='xy', scale=1, color='pink', label='vector_scalar_multiplication')


# Set limits and add grid, title, and legend
ax.set_xlim([-10, 10])
ax.set_ylim([-10, 10])
ax.grid()
ax.set_title('Vector Visualization and Operations')
ax.legend()

# Display the plot
plt.show()
```
